```python
# JPMorgan Chase
JPM = data.DataReader("JPM", 'av-daily', start, end,api_key = '3PKSOSI44CTDQYN5')
# Morgan Stanley
MS = data.DataReader("MS", 'av-daily', start, end,api_key = '3PKSOSI44CTDQYN5')
# Wells Fargo
WFC = data.DataReader("WFC", 'av-daily', start, end,api_key = '3PKSOSI44CTDQYN5')
```

** Create a list of the ticker symbols (as strings) in alphabetical order. Call this list: tickers**

```python
tickers = ['BAC', 'C', 'GS', 'JPM', 'MS', 'WFC']
```

```python
bank_stocks = pd.concat([BAC,C,GS,JPM,MS,WFC],axis =1,keys=tickers)
bank_stocks
```

```python
bank_stocks.columns.names = ['Bank Ticker','Stock Info']
```

```python
bank_stocks.head()
```

# EDA



```python
bank_stocks.xs(key='close',axis=1,level = 'Stock Info' ).max()
```

```python
returns = pd.DataFrame()
```

```python
for tick in tickers:
    returns[tick+' Return'] = bank_stocks[tick]['close'].pct_change()
returns.head()
```

```python
returns[1:]
```

```python
sns.pairplot(returns[1:])
```

```python
# best day returns
returns.idxmax()
```

```python
# worst day returns
returns.idxmin()
```

```python
returns.std()
```

```python
returns['2015-01-01': ].std()
```

```python
sns.distplot(returns['2015-01-01':]['MS Return'],color='blue',bins=100)
```

** Create a distplot using seaborn of the 2008 returns for CitiGroup **

```python
sns.distplot(returns['2008-01-01':'2009-01-01']['C Return'],color='red',bins=100)
```

# More Visualization



```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline

import plotly
import cufflinks as cf
cf.go_offline()
```

```python
for tick in tickers:
    bank_stocks[tick]['close'].plot(figsize=(20,20),label=tick)
plt.legend()
```

```python
bank_stocks.xs(key='close',axis=1,level='Stock Info').iplot()
```

## Moving Averages



```python
plt.figure(figsize=(20,20))
BAC['close']['2008-01-01':'2009-01-01'].rolling(window=30).mean().plot(label='30 Day Avg')
BAC['close']['2008-01-01':'2009-01-01'].plot(label='BAC CLOSE')
plt.legend()
```

```python
sns.heatmap(bank_stocks.xs(key='close',axis=1,level='Stock Info').corr(),annot=True)
```

```python
sns.clustermap(bank_stocks.xs(key='close',axis=1,level='Stock Info').corr(),annot=True)
```
